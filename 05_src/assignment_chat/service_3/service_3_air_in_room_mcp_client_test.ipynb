```python
from openai import OpenAI

tools = [
    {
        "type": "mcp",
        "server_label": "air_measurements_in_room_service",
        "server_description": "Provides room air measurements.",
        "server_url": mcp_url,
        "require_approval": "never",
    },
]


client = OpenAI(base_url='https://k7uffyg03f.execute-api.us-east-1.amazonaws.com/prod/openai/v1', 
                api_key='any value',
                default_headers={"x-api-key": os.getenv('API_GATEWAY_KEY')})

resp = client.responses.create(
    model="gpt-4o-mini", #"gpt-5",
    tools=tools,
    instructions="Use the air measurements service to answer the question.",
    input="Provide room readings in living room.",
)


# print(resp)
print(resp.output_text)
```
